```python
from ncclient import manager
import xmltodict
import json

HOST = 'r1.udx'
PORT = 8321
USER = 'admin'
PASS = 'admin'
name = 'default'


def get(HOST, PORT, USER, PASS):
    with manager.connect(host=HOST, port=PORT, username=USER, password=PASS,
                         hostkey_verify=False, device_params={'name': name},
                         look_for_keys=False, allow_agent=False) as m:
        return m.get(source='running').data_xml


def get_config(HOST, PORT, USER, PASS):
    with manager.connect(host=HOST, port=PORT, username=USER, password=PASS,
                         hostkey_verify=False, device_params={'name': name},
                         look_for_keys=False, allow_agent=False) as m:
        return m.get_config(source='running').data_xml
    
    
def get_capabilities(HOST, PORT, USER, PASS):

    with manager.connect(host=HOST, port=PORT, username=USER, password=PASS,
                         hostkey_verify=False, device_params={'name': name},
                         look_for_keys=False, allow_agent=False) as m:
        return m.server_capabilities
```

```python
capabilitis = get_capabilities(HOST, PORT, USER, PASS)
```

```python
running_xml = get_config(HOST, PORT, USER, PASS)
```

```python
for capability in capabilitis:
    print(capability)
```

```python
running_dict = xmltodict.parse(running_xml)
running_json = json.dumps(running_dict, indent=4)
```
